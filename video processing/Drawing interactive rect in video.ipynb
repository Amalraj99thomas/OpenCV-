```python
import cv2
    
#callback function
def draw_rect(event,x,y,flags,param):
    
    global p1,p2,topLeft_clicked,botRight_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        #reset the rectangle
        if topLeft_clicked == True and botRight_clicked == True:
            
            topLeft_clicked == False
            botRight_clicked == False
            
            p1 = (0,0)
            p2 = (0,0)

            
        if topLeft_clicked == False:
            p1 = (x,y)
            topLeft_clicked = True
            
        elif botRight_clicked == False:
            p2 = (x,y)
            botRight_clicked = True



#global variables

#initiatizing points for the rectangle
p1 = (0,0)
p2 = (0,0)

#initializing trackers for implementation
topLeft_clicked = False
botRight_clicked = False

##connect frame to the function
cap = cv2.VideoCapture(0) 

cv2.namedWindow('image')

cv2.setMouseCallback('image', draw_rect)


while True: 
    
    ret, frame = cap.read()
    
    if topLeft_clicked: #boolean so no need to write == True
        cv2.circle(frame, center=p1, radius=3, color=(0,0,255), thickness=-1)
        
    if topLeft_clicked and botRight_clicked: 
        cv2.rectangle(frame,p1,p2, (0,0,255), 2)
        
    
    #Display the resulting frame
    cv2.imshow('image', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

#release the capture once process is complete        
cap.release()
cv2.destroyAllWindows()
```
