```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
flat = cv2.imread('flat_chessboard.png')
flat = cv2.cvtColor(flat, cv2.COLOR_BGR2RGB)
plt.imshow(flat)
```

```python
flat_g = cv2.cvtColor(flat, cv2.COLOR_RGB2GRAY)
plt.imshow(flat_g, cmap='gray')
```

```python
real = cv2.imread('real_chessboard.jpg')
real = cv2.cvtColor(real, cv2.COLOR_BGR2RGB)
plt.imshow(real)
```

```python
real_g = cv2.cvtColor(real, cv2.COLOR_RGB2GRAY)
plt.imshow(real_g, cmap='gray')
```

```python
#HARRIS CORNER DETECTION
```

```python
# Convert Gray Scale Image to Float Values
gray = np.float32(flat_g)

dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)

# result is dilated for marking the corners, not important to actual corner detection
# this is just so we can plot out the points on the image shown
dst = cv2.dilate(dst,None)

# Threshold for an optimal value, it may vary depending on the image.
flat[dst>0.01*dst.max()] = [255,0,0]

plt.imshow(flat)
```

```python
gray = np.float32(real_g)

dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)

dst = cv2.dilate(dst,None)

real[dst>0.01*dst.max()]=[255,0,0]

plt.imshow(real)
```

#Shi Tomasi technique, based on a quality parameter 

```python
real = cv2.imread('real_chessboard.jpg')
real_g = cv2.cvtColor(real, cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(real_g,64,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y =i.ravel()
    cv2.circle(real, (x,y),3,(255,0,0),-1)
    
```

```python
plt.imshow(real)
```
