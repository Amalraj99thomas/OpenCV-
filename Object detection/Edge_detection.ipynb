```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
face = cv2.imread('sammy_face.jpg')
plt.imshow(face)
```

```python
#edges = cv2.Canny(image=face, threshold1=127, threshold2=127)
edges = cv2.Canny(image=face, threshold1=34, threshold2=200)
plt.imshow(edges)
```

```python
#Equation for deciding thresholds
```

```python
med_val = np.median(face)
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255, 1.3*med_val))
```

```python
edges = cv2.Canny(image=face, threshold1=lower, threshold2=upper+200) #adding 200 generalizes the edges better 
plt.imshow(edges)
```

 #image is bit dull, therefore the upper value is low. 
    Hence we can apply blurring before using canny 

```python
blurred = cv2.blur(face, ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred, threshold1=lower, threshold2=upper+30)
plt.imshow(edges)
```
