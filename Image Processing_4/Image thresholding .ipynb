```python
#Thresholding in real world applications
```

```python
img = cv2.imread('crossword.jpg',0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
#127 is the threshold value, 255 is the absolute max value

ret, img2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
show_pic(img2)
```

```python
#Adaptive thresholding
```

```python
img3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(img3)
```

```python
#we can blend both to get a better result: 

blended_img = cv2.addWeighted(src1=img2, alpha = 0.4, src2=img3, beta=0.6, gamma=0)
show_pic(blended_img)
```
