```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img = cv2.imread('puppy.jpg')
```

```python
plt.imshow(img)
```

```python
#converting image from BGR to HSV(Hue,saturation, value)

fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(fixed_img)
```

```python
#converting image from BGR to HLS(Hue,Lightness, saturation)

fixed_im = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(fixed_im)
```
