```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

#if we use OpenCV to calculate anything channel oriented, we'll want to make sure we keep the original RGB ordering, however if we ever want to display the image, then we'll need to convert to the RGB ordering that matplotlib wants.

```python
dark_horse = cv2.imread('horse.jpg') #ORIGINAL BGR OPENCV 
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB) # CONVERTED TO RGB FOR SHOW

rainbow = cv2.imread('rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
hist_values = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
img = blue_bricks
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Blue Bricks Image')
plt.show()
```
